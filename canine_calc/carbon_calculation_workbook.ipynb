```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Set up GHG emission rates for various food sources

```python
# Add up estimations of land-use, freshwater consumption, and direct GHG emissions
# Then normalize to get tons CO2e per kcal
pork = (12.)    / 1e6
eggs = (18.)    / 1e6
fish = (25.)   / 1e6
poultry = (30.) / 1e6
dairy = (38.)   / 1e6
beef = (220.) / 1e6 
```

```python
# Read nutrient stats file
food_brand_stats = pd.read_csv("static/canine_calc/Food_Stats.csv", usecols=[0, 1, 2, 4, 5, 6, 7, 8, 9, 10])
print(tuple([tuple((x,x)) for x in food_brand_stats["Brand"].values]))
chart = sns.barplot(data=food_brand_stats, y="Brand", x="Poultry")
# sns.barplot(x=food_brand_stats["Brand"], y=food_brand_stats["Poultry"])
```

```python
import cmocean
# Multiply food source percentages by amount consumed by GHG emission rates

# User input
key = "IAMS Proactive Health (Lamb & Rice)"
daily_cups = 2

# Siphon stats by brand
brand_stats = food_brand_stats[food_brand_stats.Brand == key]

# kCals per cup attributable to each type of animal-based food source
brand_protein_amts = pd.DataFrame(brand_stats[["Pork", "Eggs", "Fish", "Poultry", "Dairy", "Beef"]] * \
                    brand_stats[["kCal per cup"]].values)
brand_protein_amts.index = [key]

# Get total daily amounts and calculate GHG emissions per day
emissions = brand_protein_amts[["Pork", "Eggs", "Fish", "Poultry", "Dairy", "Beef"]] * daily_cups * \
                  np.hstack((pork, eggs, fish, poultry, dairy, beef))
emissions_annual = emissions * 365. # get annual GHG emissions
print(emissions_annual)

# Get emissions for all brands
for i, food_brand in enumerate(food_brand_stats["Brand"]):
    brand_stats = food_brand_stats[food_brand_stats.Brand == food_brand]
    tmp_amts = pd.DataFrame(brand_stats[["Pork", "Eggs", "Fish", "Poultry", "Dairy", "Beef"]] * \
                    brand_stats[["kCal per cup"]].values)
    print(tmp_amts)
    if i == 0:
        all_protein_amts = tmp_amts
    else:
        all_protein_amts = pd.concat([all_protein_amts, tmp_amts])
all_protein_amts.index = [food_brand for food_brand in food_brand_stats["Brand"]]
all_emissions = all_protein_amts * daily_cups *  np.hstack((pork, eggs, fish, poultry, dairy, beef))
all_annual_emissions = all_emissions * 365.
inds = all_annual_emissions.sum(axis=1).argsort()
print(all_annual_emissions.sum(axis=1))

# Plot
fig = plt.figure(figsize=(10,12))

sns.set_palette('cmo.matter'); sns.set_context("paper"); sns.set_style("darkgrid")
cs=(cmocean.cm.matter(np.arange(100)/4.))

ax1 = plt.subplot(221)
sns.barplot(data=emissions_annual, ax=ax1); ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title(f"Annual Emissions by Protein Type\nBrand: {key} ({daily_cups} cups/day)", fontweight='bold')
ax1.set_xlabel("Protein"); ax1.set_ylabel(r"Emissions (tons CO$_{2}$e)")

ax2 = plt.subplot(222)
ax2.pie(emissions_annual.T.values[np.where(emissions_annual.T.values != 0)],
        labels=emissions_annual.T.index.values[np.where(emissions_annual.T.values != 0)[0]], 
        autopct='%1.1f%%', startangle=90, colors=cs, shadow=True)
ax2.set_title(f"Annual Emissions Breakdown by Protein\nBrand: {key} ({daily_cups} cups/day)", fontweight='bold')

ax3 = plt.subplot(212)
cs=(cmocean.cm.amp(np.arange(2, 12)/15))
sns.barplot(x=all_annual_emissions.index[inds], y=all_annual_emissions.sum(axis=1)[inds], ax=ax3, 
            palette=np.asarray([cs[i] if (key not in food_brand_stats.Brand[i]) \
                              else "#4e3799" for i in range(len(all_annual_emissions.index))]),
            edgecolor='darkgrey')
ax3.set_title(f"Ranked Annual Carbon Emissions by Brand (given {daily_cups} cups/day)", fontweight='bold')
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=90)
```
